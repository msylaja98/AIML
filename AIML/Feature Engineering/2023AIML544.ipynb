* Parental Support : Categorical (Ordinal)

We can handle missing data in multiple ways. 

For numerical attribute, we can use mean/median imputation, random sampling imputation, adding a new variable to indicate missingness, imputation of missing values by the values at the end of distribution or imputation of NA by arbitrary values.

For categorical attribute, we can use mode imputation, median imputation for ordinal ones or treating NA as additional category etc.

### 3. Give the reasons for each column that how you are handling the data.

* Age :- Since this is a student performance data, all the age of the students will be more or less the same. So we can impute the missing values by the mean of age attribute.

```python
mean_age = df['Age'].mean()
print(mean_age)

# Replace missing values with mean
df.fillna({'Age':mean_age}, inplace=True)
```

* StudyTimeWeekly :- Since this attribute indicated the number of hours the student spend for studying per week, we can impute the missing values by mean value. Students of a same class must spend some average amount of time to complete the course.

```python
mean_study_time = df['StudyTimeWeekly'].mean()
print(mean_study_time)

# Replace missing values with mean
df.fillna({'StudyTimeWeekly':mean_study_time}, inplace=True)
```

* Absences :- Here also we can impute the missing values by mean because the number of absences of a student will be same on an average.

```python
mean_absence = df['Absences'].mean()
print(mean_absence)

# Replace missing values with mean
df.fillna({'Absences':mean_absence}, inplace=True)
```

* Tutoring :- Since this is a categorical nominal (binary) attribute we can use mode to impute the missing value.

```python
mode_tutoring = df['Tutoring'].mode().iloc[0]
print(mode_tutoring)

# Replace missing values with mode
df.fillna({'Tutoring':mode_tutoring}, inplace=True)
```

* ParentalSupport :- Since this is a categorical ordinal attribute we can use median to impute the missing value.

```python
median_parental_support = df['ParentalSupport'].median()
print(median_parental_support)

# Replace missing values with median
df.fillna({'ParentalSupport':median_parental_support}, inplace=True)
```

* Sports :- Since this is a categorical nominal (binary) attribute we can use mode to impute the missing value.

```python
mode_sports = df['Sports'].mode().iloc[0]
print(mode_sports)

# Replace missing values with mode
df.fillna({'Sports':mode_sports}, inplace=True)
```

* Music :- Since this is a categorical nominal (binary) attribute we can use mode to impute the missing value.

```python
mode_music = df['Music'].mode().iloc[0]
print(mode_music)

# Replace missing values with mode
df.fillna({'Music':mode_music}, inplace=True)
```

### 7. Calculate the proximity measure between ( choose the measuring technique based on the column):

#### i)Tutoring and GPA
Since tutoring is a categorical(binary) attribute we can use hamming distance to create the dissimilarity matrix based on tutoring.
Since GPA is numerical attribute we can use euclidean distance to create the dissimilarity matrix based on GPA.
Then we can take an average of the each elements in the matrix to create a new dissimilarity matrix which will be proximity measure based on tutoring and GPA.

```python
import numpy as np
from sklearn.metrics import pairwise_distances

# Select relevant column for computing dissimilarity
X = df[['Tutoring']].values

# Calculate pairwise Hamming distances
dissimilarity_matrix_on_tutoring = pairwise_distances(X, metric='hamming')

# Select relevant column for computing dissimilarity
X = df[['GPA']].values

# Calculate pairwise Euclidean distances
dissimilarity_matrix_on_gpa = pairwise_distances(X, metric='euclidean')

dissimilarity_matrix_on_tutoring_and_gpa = (np.array(dissimilarity_matrix_on_tutoring) + np.array(dissimilarity_matrix_on_gpa))/2
print(dissimilarity_matrix_on_tutoring_and_gpa)
```

#### ii) StudyTime and GPA

Since both studytime and GPA are numerical attributes we can use euclidean distance to create the dissimilarity matrix.

```python
# Select relevant column for computing dissimilarity
X = df[['StudyTimeWeekly','GPA']].values

# Calculate pairwise Euclidean distances
dissimilarity_matrix_on_studytime_and_gpa = pairwise_distances(X, metric='euclidean')

print(dissimilarity_matrix_on_studytime_and_gpa)
```

#### iii) Studytime and absences

Since both studytime and absences are numerical attributes we can use euclidean distance to create the dissimilarity matrix.

```python
# Select relevant column for computing dissimilarity
X = df[['StudyTimeWeekly','Absences']].values

# Calculate pairwise Euclidean distances
dissimilarity_matrix_on_studytime_and_absences = pairwise_distances(X, metric='euclidean')

print(dissimilarity_matrix_on_studytime_and_absences)
```

### 8. Perform the feature selection considering GPA as the target variable and the rest of the columns as independent variables using the filter method.  ( use at least 3 methods )

**For the independant numerical attributes Age, StudyTime, Absences we can use pearson correlation coefficient to identify the correlation with GPA** 

```python
from scipy.stats import pearsonr
corr_coef_age, p_value = pearsonr(df['Age'], df['GPA'])
print('{:f}'.format(corr_coef_age))
```

**There is no proper linear relation between Age and GPA**

```python
corr_coef_studytime, p_value = pearsonr(df['StudyTimeWeekly'], df['GPA'])
print(corr_coef_studytime)
```

**There is a positive linear relation between StudyTimeWeekly and GPA**

```python
corr_coef_absences, p_value = pearsonr(df['Absences'], df['GPA'])
print(corr_coef_absences)
```

**There is a highly negative linear relation between Absences and GPA**

**In order to do feature selection for categorical attributes we need to first discretize GPA into categorical bins or category**

```python
from sklearn.preprocessing import KBinsDiscretizer
discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')  # Adjust n_bins as needed
df['GPA_bins'] = discretizer.fit_transform(df[['GPA']])
```

```python
print(df['GPA_bins'])
```

**Now the GPA has been discretized.**

**Chi square test on categorical attributes and GPA_bins**

```python
from sklearn.feature_selection import chi2
categorical_cols = ['Gender', 'Ethnicity', 'ParentalEducation', 'Tutoring', 'ParentalSupport', 'Extracurricular', 'Sports', 'Music', 'Volunteering','GradeClass']

# Calculate chi-square scores and p-values
chi2_scores, p_values = chi2(df[categorical_cols], df['GPA_bins'])

# Set NumPy print options to suppress scientific notation
pd.set_option('display.float_format', lambda x: '%.3f' % x)

# Display chi-square scores and p-values for each feature
chi2_results = pd.DataFrame({
    'Feature': categorical_cols,
    'Chi2 Score': chi2_scores,
    'P-value': p_values
})

print("Chi-square Test Results:")
print(chi2_results)
```

**Feature selection using information gain on categorical attributes and GPA_bins**

```python
from sklearn.feature_selection import mutual_info_classif
mi_scores = mutual_info_classif(df[categorical_cols], df['GPA_bins'])

# Display mutual information or information gain scores for each feature
mi_scores_df = pd.DataFrame(data=mi_scores, index=categorical_cols, columns=['Mutual Information Score'])
print("Mutual Information Scores:")
print(mi_scores_df)
```
